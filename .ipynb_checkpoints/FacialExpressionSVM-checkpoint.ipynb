```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import utils,numpy as np
from skimage.io import *
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, cross_val_score, cross_val_predict
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import pickle
```

```python
a= utils.load_training_images_features()
y= utils.get_training_labels()
```

```python
sss = StratifiedShuffleSplit(n_splits=1,test_size=0.4)
for train_index,test_index in sss.split(a,y):
    x_train,x_test = a[train_index],a[test_index]
    y_train,y_test = y[train_index],y[test_index]
```

```python
scaler=StandardScaler()
pca = PCA(n_components=0.999)
svc = SVC(probability=True,kernel="linear",C=1)
bagging = BaggingClassifier(svc,oob_score=True)
```

```python
pipe = Pipeline(steps=[
    ("scaler",scaler),
    #("pca",pca),
    ("svc",svc)
])
```

```python
pipe.fit(x_train,y_train)
```

```python
cross_val_score(pipe,x_train,y_train,cv=5)
```
